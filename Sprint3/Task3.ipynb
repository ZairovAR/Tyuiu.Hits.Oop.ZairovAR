<h2 style="color:DodgerBlue">Описание проекта:</h2>

Создать базовый класс Book в C#, который будет представлять информацию о книгах. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для демонстрации полиморфизма. 

Требования к базовому классу Book: 

• Атрибуты: Название (Title), Автор (Author), Год издания (YearOfPublication). 

• Методы: o GetInfo(): метод для получения информации о книге в виде строки. 

o Read(): метод для вывода сообщения о чтении книги. 

o Borrow(): метод для вывода сообщения о выдаче книги на чтение.


Требования к производным классам: 
1. Учебник (Textbook): Должен содержать дополнительные атрибуты, такие как Предмет (Subject). Метод Read() должен быть переопределен для добавления информации о предмете при чтении учебника. 

2. Художественная литература (Fiction): Должен содержать дополнительные атрибуты, такие как Жанр (Genre). Метод Borrow() должен быть переопределен для добавления информации о жанре при выдаче книги на чтение. 

3. Научная литература (ScientificLiterature) (если требуется третий класс): Должен содержать дополнительные атрибуты, такие как Область науки (FieldOfScience). Метод GetInfo() должен быть переопределен для включения информации об области науки в описании книги. 

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

```
public class Book
{
    protected string Title { get; set; }
    protected string Author { get; set; }
    protected int YearOfPublication { get; set; }

    public Book(string title, string author, int yearOfPublication)
    {
        this.Title = title;
        this.Author = author;
        this.YearOfPublication = yearOfPublication;
    }

    public virtual string Borrow()
    {
        return "Вы взяли книгу.";
    }
}

public interface IGetInfo
{
    string GetInfo();
}

public interface IRead
{
    string Read();
}

public interface IBorrow
{
    string Borrow();
}

class Textbook : Book, IRead, IBorrow, IGetInfo
{
    protected string Subject { get; set; }

    public Textbook(string title, string author, int yearOfPublication, string subject)
        : base(title, author, yearOfPublication)
    {
        this.Subject = subject;
    }

    public string GetInfo()
    {
        return $"Название книги: {Title}\n Автор: {Author}\n Год издания: {YearOfPublication}\n Предмет: {Subject} \n";
    }

    public string Read()
    {
        return $"Вы читаете учебник про {Title}.";
    }

    public override string Borrow()
    {
        return "Вы взяли учебник.";
    }

    public string Borrow(string student)
    {
        return $"{student} взял учебник по предмету {Subject}.";
    }
}

class Fiction : Book, IBorrow, IRead, IGetInfo
{
    protected string Genre { get; set; }

    public Fiction(string title, string author, int yearOfPublication, string genre)
        : base(title, author, yearOfPublication)
    {
        this.Genre = genre;
    }

    public string GetInfo()
    {
        return $"Название книги: {Title}\n Автор: {Author}\n Год издания: {YearOfPublication}\n Жанр: {Genre} \n";
    }

    public string Read()
    {
        return $"Вы читаете книгу в жанре {Genre}.";
    }

    public override string Borrow()
    {
        return "Вы взяли художественную книгу.";
    }

    public string Borrow(string readerType)
    {
        return $"{readerType} взял художественную книгу.";
    }
}

class ScientificLiterature : Book, IGetInfo, IBorrow, IRead
{
    protected string FieldOfScience { get; set; }

    public ScientificLiterature(string title, string author, int yearOfPublication, string fieldOfScience)
        : base(title, author, yearOfPublication)
    {
        this.FieldOfScience = fieldOfScience;
    }

    public string GetInfo()
    {
        return $"Название книги: {Title}\n Автор: {Author}\n Год издания: {YearOfPublication}\n Область науки: {FieldOfScience} \n";
    }

    public string Read()
    {
        return $"Вы читаете книгу области науки {FieldOfScience}.";
    }

    public override string Borrow()
    {
        return "Вы взяли научную книгу.";
    }

    
    public string Borrow(string researcher)
    {
        return $"{researcher} взял научную книгу по {FieldOfScience}.";
    }
}


public class Library<T> where T : Book
{
    private List<T> _books = new List<T>();

    
    public void AddBook(T book)
    {
        _books.Add(book);
        Console.WriteLine("Книга добавлена в библиотеку.");
    }

    
    public void DisplayBooks()
    {
        foreach (var book in _books)
        {
            Console.WriteLine("\nКниги в библиотеке: \n");
            Console.WriteLine(book is IGetInfo info ? info.GetInfo() : "Информация недоступна.");
        }
    }
}


     
//Fiction fictionBook = new Fiction("Ауук", "Афанасьев Лисьев", 1934, "Драма");
Textbook textbook = new Textbook("Матричные уравнения", "Иванов", 2021, "Математика");
//ScientificLiterature scienceBook = new ScientificLiterature("Наука о жизни", "Петров", 2018, "Биология");

       
//Console.WriteLine(fictionBook.GetInfo());
//Console.WriteLine(fictionBook.Read());
//Console.WriteLine(fictionBook.Borrow("Студент"));

Console.WriteLine(textbook.GetInfo());
Console.WriteLine(textbook.Read());
Console.WriteLine(textbook.Borrow("Учитель"));

//Console.WriteLine(scienceBook.GetInfo());
//Console.WriteLine(scienceBook.Read());
//Console.WriteLine(scienceBook.Borrow("Исследователь"));


Library<Book> library = new Library<Book>();
//library.AddBook(fictionBook);
library.AddBook(textbook);
//library.AddBook(scienceBook);


library.DisplayBooks();
```
